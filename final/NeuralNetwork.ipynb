```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

train_dataset = pd.read_csv('Training.csv')
test_dataset = pd.read_csv('Testing.csv')
X = train_dataset.iloc[:, 1:132].values
y = train_dataset.iloc[:, 132].values
X
```

```
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_y_1 = LabelEncoder()
y = labelencoder_y_1.fit_transform(y)
y





```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
y_train

```

```


#importing the keras libraries and keras
import keras
from keras.models import Sequential
from keras.layers import Dense

#initializing the ANN
classifier = Sequential()

#adding an input layer and the first hidden layer
classifier.add(Dense(54, input_shape =(131,), activation = 'relu'))


classifier.add(Dense(108, activation = 'relu'))

classifier.add(Dense(49, activation = 'softmax'))

classifier.summary()
classifier.compile(optimizer = 'adam' , loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train,y_train,batch_size = 10 ,epochs=20)
_, accuracy = classifier.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```
predictions = classifier.predict_classes(X_test)
# summarize the first 5 cases
for i in range(30):
	print('%s => %d (expected %d)' % (X_test[i].tolist(), predictions[i], y_test[i]))
```
